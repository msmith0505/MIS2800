Question 1: By enabling Lyft to examine incredibly enormous volumes of ride and customer data far more quickly than traditional databases, Amazon Redshift contributed to the company's innovation. Question 2: Because each system is made for a different purpose, companies usually don't utilize the same database for both OLTP and OLAP workloads. OLTP systems concentrate on quick, real-time transactions where accuracy and speed are crucial, like scheduling transportation or handling payments. OLAP systems, such as Amazon Redshift, are designed to analyze big datasets and execute intricate queries for business intelligence and reporting. Running analytical queries on the same database that handles transactions could cause operations to lag and have a detrimental effect on users. Businesses may preserve performance and still obtain insightful information from their data by separating the systems.

## YOUR ANSWERS TO PART 2 BELOW

Question 1: Because data is stored in columns rather than rows, columnar storage enhances analytical performance. Analytics searches frequently simply require individual columns rather than complete records. Amazon Redshift speeds up and improves the efficiency of queries by reading only the necessary columns, which minimizes the quantity of data that needs to be scanned. For reporting, aggregation, and large-scale data analysis, columnar storage is more appropriate than the conventional normalized row-based databases that were covered earlier in the course.

Question 2: Scalability and distributed computing are two ways that Amazon Redshift enhances performance. Data is dispersed among several nodes that cooperate to perform queries concurrently. Large datasets can be swiftly evaluated because to this parallel processing. Redshift is efficient and economical for big businesses managing enormous volumes of data because it runs in the AWS cloud, which allows businesses to scale processing and storage resources up or down based on their needs.
